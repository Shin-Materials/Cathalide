## 1. Rotating a molecule in the structure

### Required variables are as follows:
- Define molecules with VESTA labels
- Set the anchor point (or reference point) of rotation
- Set the axis of rotation (numpy array of [3,]). Direction is based on a,b,c vectors of POSCAR file
- Set degree of rotation

```python
%matplotlib inline
from pymatgen.core import Structure
from pymatgen.core import Element
from glob import glob
import numpy as np
from scipy.spatial.transform import Rotation as R
from Functions import molecule_rotation, convert_site_index, Write_POSCAR
from Functions import list_all_molecules, molecule_finder, create_df
from Functions import Shin_molecule
import copy
import pandas as pd
import scipy

"""
1. Import POSCAR structure as pymatgen structure
"""
struct=Structure.from_file("FACsPb2Br6.vasp")

"""
2. Define molecules by listing VESTA labels
"""
molecule = ['C2', 'N6', 'H2', 'N2', 'H14', 'H18', 'H10', 'H6']

### Note that you can also idenfity this by using 'molecule finder' function
# molecule = molecule_finder(struct,'C2')

"""
3. Set values needed for rotation (anchor point, axis_vector, angle_degree)
"""
# Anchor point, here I set it to be the position of 'C2'
reference_point = np.array([0.25000, 0.7877, 0.25])
### If you want to identify the position of certain atom, you can use the followings:
# df=create_df(struct)
# reference_point=(df[df['atom_label']=='C2']['pmg_site'].iloc[0]).frac_coords

# Vector of rotation axis
axis_vector = np.array([0,1,0])
# Angle, assuming counter clock-wise
angle_degree = 90 #degree

"""
4. Rotation and write to POSCAR
"""
rotated_struct = molecule_rotation(struct,molecule,axis_vector,angle_degree, reference_point)
Write_POSCAR(rotated_struct,filename='Rotated_molecule_C2.vasp')
```

```python
"""
THIS CELL IS ONLY TO VISUALIZE THE BEFORE AND AFTER
"""
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img_before=mpimg.imread('./Test_structures/FACsPb2Br6_before_rotation.png')
img_after=mpimg.imread('./Test_structures/FACsPb2Br6_after_rotation.png')
fig, ax = plt.subplots(1,2,figsize=(10,4.5))
plt.tight_layout()
ax[0].imshow(img_before)
ax[1].imshow(img_after)
ax[0].axis('off')
ax[1].axis('off')
circle1 = plt.Circle((230,460),80,color='r',fill=None,linewidth=4)
circle2 = plt.Circle((230,460),80,color='r',fill=None,linewidth=4)
ax[0].add_artist(circle1)
ax[1].add_artist(circle2)
ax[0].text(100,740,'BEFORE ROTATION',ha='left',fontsize=20)
ax[1].text(100,740,'AFTER ROTATION',ha='left',fontsize=20)
```
